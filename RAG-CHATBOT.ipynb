```python
# prompt: pymupdf complete query

from langchain.document_loaders import PyMuPDFLoader

loader = PyMuPDFLoader("/content/Data Science.pdf") # replace with your file
pages = loader.load()
print(len(pages))
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

def split_docs(pages, chunk_size=1000, chunk_overlap=20):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
  docs = text_splitter.split_documents(pages)
  return docs

docs = split_docs(pages)
print(len(docs))
```

```python
%%writefile utils.py
import streamlit as st
from streamlit_chat import message
import fitz
```

```python
# Load the PDF file
def load_pdf(uploaded_file):
    doc = fitz.open(uploaded_file)
    return doc
```

```python
# Extract text from the PDF
def extract_text(doc):
    text = ''
    for page in doc:
        text += page.get_text()
    return text
```

```python
# Extract answer from the text based on the question
def extract_answer(question, text):
    # Simple keyword-based search
    keywords = question.split()
    for keyword in keywords:
        if keyword in text:
            # Return the surrounding text as the answer
            start_idx = text.find(keyword)
            end_idx = start_idx + 200  # Return 200 characters surrounding the keyword
            return text[start_idx:end_idx]
    return "Answer not found"
```

```python
# Process the user input
def process_function(user_input, uploaded_file):
    if user_input:
        if uploaded_file is not None:
            doc = load_pdf(uploaded_file)
            text = extract_text(doc)
            answer = extract_answer(user_input, text)
            message(answer, key="bot")
        else:
            message("Please upload a PDF file", key="bot")
    else:
        message("Please enter your question", key="bot")
```

```python
# Streamlit app
import streamlit as st
def process_function(user_input, uploaded_file):
  st.title("Welcome to Rag Chat-bot")
  uploaded_file = st.file_uploader("Choose a PDF file")
  user_input = st.text_input("Enter your Question:")
  process_function(user_input, uploaded_file)
```
